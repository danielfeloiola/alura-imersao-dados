```python
import pandas as pd
```

```python
url = 'https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv'
dados = pd.read_csv(url)
dados.head()
```

## Desafio 1 - Média da metragem por bairro

```python
dados.groupby(['Bairro'])['Metragem'].mean()
```

## Desafio 2 - Duas formas de selecionar os dados por bairro 

```python
dados.groupby('Bairro').mean()
```

## Desafio 3 - Explorar gráficos

```python
dados.groupby(['Bairro'])['Metragem'].mean().sort_values(ascending=False).head().plot()
```

O bairro com maior metragem é o Jardim Represa LOL



## Desafio 4 - Testar média, mediana, min, max


```python
dados['Metragem'].mean(), dados['Metragem'].median(), dados['Metragem'].min(), dados['Metragem'].max()
```

## Desafio 5 - Quais bairros não tem nome de rua?

```python
dados[pd.isna(dados["Rua"])].groupby("Bairro")["Bairro"].count()
```
