```python
valor_por_bairro = dados_vendas.groupby('Bairro')['Valor_anuncio_float'].sum()
metragem_dos_bairros =  dados_vendas.groupby('Bairro')['Metragem'].sum()


valor_metro_quadrado = valor_por_bairro/metragem_dos_bairros
valor_metro_quadrado = valor_metro_quadrado.sort_values(ascending=False)
plt.figure(figsize=(20, 10))
valor_metro_quadrado[:30].plot(kind='bar')

plt.title('Valor do metro quadrado dos bairros mais caros de SP')
plt.ylabel('Valor do metro quadrado')
plt.show()
```
